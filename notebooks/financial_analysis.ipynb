# 💹 Financial Performance Analysis Dashboard

Financial KPI analysis using Python, SQL, and Power BI.

```python
import pandas as pd

df = pd.read_csv(
    "../data/financial_data.csv"
)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Financial KPIs

```python
total_revenue = df["Revenue"].sum()

total_expenses = df["Expenses"].sum()

total_profit = df["Profit"].sum()

avg_roi = df["ROI"].mean()

print("Revenue:", total_revenue)
print("Expenses:", total_expenses)
print("Profit:", total_profit)
print("Average ROI:", round(avg_roi,2))
```

```python
import matplotlib.pyplot as plt

monthly_revenue = (
    df.groupby("Month")
    ["Revenue"]
    .sum()
)

monthly_revenue.plot(
    figsize=(12,5)
)

plt.title(
    "Monthly Revenue Trend"
)

plt.show()
```

```python
department_profit = (
    df.groupby("Department")
    ["Profit"]
    .sum()
)

department_profit.plot(
    kind="bar",
    figsize=(8,4)
)

plt.title(
    "Profit by Department"
)

plt.show()
```

# Key Insights

- Revenue increased steadily across the reporting period.
- Sales generated the highest overall profit.
- Marketing produced strong ROI despite lower revenue.
- Certain departments may require expense optimization.
